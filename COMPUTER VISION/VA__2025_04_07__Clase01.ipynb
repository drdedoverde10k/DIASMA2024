```python
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
```

```python
# Load the image
img = io.imread("https://raw.githubusercontent.com/jorgevergara/PDI/refs/heads/main/images/aero.jpg")
img_gray = io.imread("https://raw.githubusercontent.com/jorgevergara/PDI/refs/heads/main/images/aero.jpg",as_gray=True)
```

```python
plt.figure()
plt.imshow(img[300:450:2,400:200:-1,[2,1,0]])
```

```python
plt.figure()
plt.imshow(np.log(1+(1-img_gray)),cmap='binary')
```

```python
plt.figure()
plt.imshow(img_gray>0.3,cmap='gray')
```
