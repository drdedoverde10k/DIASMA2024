```python
import numpy as np
import sympy as sp
from sympy import * # This imports all the functions and classes from SymPy into our interactive Python session. Now, suppose we start to do a computation.
from scipy.linalg import solve
from scipy.optimize import minimize
```

```python
#Simpy easily allows us to solve systems of equations
x1, x2, l = sp.symbols("x1 x2 l")

f = 0.25*x1**2+10*x1+0.15*x2**2+12*x2
g = x1+x2-1000
```

```python
f
```

```python
F = sp.hessian(f, (x1, x2))
F
```

```python
F.det() # Si el determinante de la matriz Hessiana es positivo, existe un mínimo. Para un problema sin restricciones.
```

Para todo valor de x1, x2 en la Matriz Hessiana de F es mayor o igual que cero, entonces la función de Costo es definida positiva y convexa. Esto significa que cualquier punto en el que las derivadas parciales son iguales a cero, son mínimos locales.

```python
g
```

```python
L = f - l*g
L
```

```python
H = sp.hessian(L, (x1, x2, l))
print("Matriz Hessiana:")
sp.pprint(H)
```

```python
H.det()
```

```python
Lx1 = diff(L, x1)
Lx1
```

```python
Lx2 = diff(L, x2)
Lx2
```

```python
Ll = diff(L, l)
```

```python
# Por lo tante tenemos las siguientes derivadas parciales, las cuales igualaremos a 0.
print(f" ∂L/∂x1 = {Lx1} = 0")
print(f" ∂L/∂x2 = {Lx2} = 0")
print(f" ∂L/∂l = {Ll} = 0")
```

```python
A = np.array([[0.5, 0, -1],
              [0, 0.3, -1],
              [-1, -2, 0]])

b = np.array([-10,-12,-1000])

sol = solve(A, b)

sol
```

Por lo tanto, la función de costo tiene un mínimo local en x1 = 233.85 y x2=383.07.

## Por lo tanto, la cantidad de bancos de Madera que deben producirse en el lugar x1 y x2 es 233 y 283 respectivamente.

```python
# Definimos las variables del sistema
x, y, l = sp.symbols("x, y, l")

f = 4*x + x*y + 2*y
g = 20*x + 4*y + -2000
```

```python
f
```

```python
g
```

```python
F = sp.hessian(f, (x, y))
F
```

```python
F.det()
```

Cómo la matriz hessiana es mayor o igual que cero para todo valor de x, y. Entonces, esto significa que cualquier punto en el que las derivadas parciales son iguales a cero, son mínimos locales.

```python
L = f - l * g
L
```

```python
R = sp.hessian(L, (x, y, l))
```

```python
R.det()
```

```python
Lx = diff(L, x)
Lx
```

```python
Ly = diff(L, y)
Ly
```

```python
Ll = diff(L, l)
Ll
```

```python
# Por lo tante tenemos las siguientes derivadas parciales, las cuales igualaremos a 0.
print(f" ∂L/∂x = {Lx} = 0")
print(f" ∂L/∂y = {Ly} = 0")
print(f" ∂L/∂l = {Ll} = 0")
```
